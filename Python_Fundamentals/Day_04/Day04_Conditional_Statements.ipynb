# Mini Project

Project Name:
Student Grade Checker

Description:

Ask the user for:

• Student Name
• Course
• Score

Display:

• Grade
• Remark

Add validation so scores below 0 or above 100 are rejected.

```python
name = input("Enter student name: ")
course = input("Enter course: ")
score = int(input("Enter score: "))

if score < 0 or score > 100:
    print("Invalid score! Score must be between 0 and 100.")
else:
    if score >= 70:
        grade = "A"
        remark = "Excellent"
    elif score >= 60:
        grade = "B"
        remark = "Very Good"
    elif score >= 50:
        grade = "C"
        remark = "Good"
    elif score >= 45:
        grade = "D"
        remark = "Fair"
    else:
        grade = "F"
        remark = "Fail"

    print("\n==============================")
    print("     STUDENT GRADE REPORT")
    print("==============================")
    print(f"Name   : {name}")
    print(f"Course : {course}")
    print(f"Score  : {score}")
    print(f"Grade  : {grade}")
    print(f"Remark : {remark}")
    print("==============================")
```

#Build a BMI Calculator.

Requirements:

• Ask for weight and height.
• Calculate BMI.
• Display whether the user is Underweight, Normal, Overweight, or Obese.
• Prevent invalid inputs such as zero or negative values.

```python
weight = float(input("Enter your weight (kg): "))
height = float(input("Enter your height (m): "))

if weight <= 0 or height <= 0:
    print("Invalid input! Weight and height must be greater than zero.")
else:
    bmi = weight / (height ** 2)

    print(f"\nYour BMI is: {bmi:.2f}")

    if bmi < 18.5:
        print("Category: Underweight")
    elif bmi < 25:
        print("Category: Normal Weight")
    elif bmi < 30:
        print("Category: Overweight")
    else:
        print("Category: Obese")
```
