# 📚 What I Learnt

A loop is used to repeat a block of code multiple times without rewriting it.

Python provides two main types of loops:

1. for loop
   Used when the number of repetitions is known.

2. while loop
   Used when the number of repetitions depends on a condition.

I also learned about:

• break – Stops the loop immediately.
• continue – Skips the current iteration and continues with the next one.

Loops reduce repetitive code and are widely used in Artificial Intelligence, Machine Learning, robotics, data processing, and automation.

#💻 Examples

```python
for number in range(1, 11):
    print(number)
```

```python
name = "Yusuf"

for i in range(5):
    print(name)
```

```python
for number in range(2, 21, 2):
    print(number)
```

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

```python
for number in range(1, 11):
    if number == 6:
        break

    print(number)
```

```python
for number in range(1, 11):

    if number == 5:
        continue

    print(number)
```

# 🎯 Challenge

Build a Number Guessing Game.

Requirements

• Secret number = 7

• Ask the user to guess.

• Keep asking until they guess correctly.

• Display:

Congratulations!

when the correct number is entered.

```python
secret_number = 7

guess = 0

while guess != secret_number:
    guess = int(input("Guess the number: "))

    if guess == secret_number:
        print("🎉 Congratulations! You guessed correctly.")
    else:
        print("❌ Wrong guess. Try again.")
```

```python
secret_number = 7
attempts = 0

while True:
    guess = int(input("Guess the number: "))
    attempts += 1

    if guess == secret_number:
        print(f"🎉 Congratulations! You guessed correctly in {attempts} attempts.")
        break
    else:
        print("❌ Wrong guess. Try again.")
```

# 🧠 Key Takeaways

✓ A loop repeats code automatically.

✓ Use a for loop when the number of repetitions is known.

✓ Use a while loop when repetition depends on a condition.

✓ break exits a loop immediately.

✓ continue skips the current iteration.

✓ Loops make programs shorter, cleaner, and more efficient.
